```python
import numpy as np 
import pandas as pd 
```

```python
df = pd.DataFrame()
for i in range(7):
    j = pd.read_csv(f"data/{i+1}.csv")
    df = pd.concat([df, j])
df.head()
```

```python
df.columns
li = ['Accepts Email Marketing','Company', 'Address1', 'Address2', 'City', 'Province', 'Province Code','Country Code', "Accepts SMS Marketing","Tax Exempt",  'Total Spent',"Tags","Note"]
df = df.drop(li, axis = 1)
```

```python
df = df.dropna()
```

```python
k = []
for i in df["Total Orders"]:
    if i==0:
        k.append(i)
    else:
        k.append("NaN")
```

```python
df["od_no"] = k
```

```python
list1 = []
for i in df["Phone"]:
    summ = 0
    for j in i:
        summ+=1
    if summ==10:
        k = ((i.strip()).replace(" ", ""))
        list1.append(k)
    elif "+91" in i:
        k = ((i[3:].strip()).replace(" ", ""))
        list1.append(k)
    elif "0" in i[0]:
        k = ((i[1:].strip()).replace(" ", ""))
        list1.append(k)
    else:
        k = (i.strip()).replace(" ", "")
        list1.append(k)
```
